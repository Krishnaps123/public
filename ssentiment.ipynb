```python
import string
```

```python
punctuationfree = ''
def remove_pun(text):
  punctuationfree = ''.join([i for i in text if i not in string.punctuation])
  return punctuationfree
```

```python
data['clean_msg'] = data['tweet'].apply(lambda x : remove_pun(x))
```

tokenization

words are tokenized so sentenced are not tokenized separately


```python
import nltk
```

```python
nltk.download('punkt')
```

```python
def tokenize(text):
  word = nltk.word_tokenize(text)
  return word
```

```python
data['tokens'] = data['tweet'].apply(lambda x : tokenize(x))
```

```python
data.head()
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
print(stopwords.fileids())
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
  output = [i for i in text if i not in stopwords]
  return output
```

```python
data['no_stopwords'] = data['tokens'].apply(lambda x : remove_stopwords(x))
```

```python
data.head()
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
  stem_text = [ps.stem(word) for word in text]
  return stem_text
```

```python
data['stem_msg'] = data['no_stopwords'].apply(lambda x : stemming(x))
```
